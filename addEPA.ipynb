# Add Offensive EPA Column

The `pbp` dataframe looks at EPA from the *home* team's perspective, so negative EPA for them means the away team's offense gained positive EPA. Because we're looking at EPA *strictly* from an offensive perspective, we want to create a column that reflects that regardless if it's home or away.

```python
# Implement function that will see if each play's EPA is from the offense or defense
# TODO: Make it so it can skip checking this process for ae entire drive and/or game

def addEPA(data):
    data['offEPA'] = 0

    for index, row in data.iterrows():
        # if the team that has posession is the home team, 
        # then epa from this play is positive
        if row['posteam'] == row['home_team']:
            row['offEPA'] = row['epa'] 
        # if the team that has posession is the away team,
        # then epa from this play will be negative, 
        # so we must negate it to be positive for the offense
        else:
            row['offEPA'] = row['epa'] * -1
        data['offEPA'][index] = row['offEPA']
```

```python
# Run function for all data frames

for i in pbp:
    addEPA(pbp[i])
    pbp[i].to_csv('data/offEPA/off_epa_20' + str(i) + '.csv')

print("done")
```
